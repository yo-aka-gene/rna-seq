### Heat Map "mini" for Glioma Bulk-seq Data 
----

#### Rで描いたやつのデザインがしょぼいので作り直す
#### mini: 既に描いてあるHeat Mapと同じ方法でより遺伝子を絞ってHeat Map作成

```python
# Import Library

import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy.stats import zscore
```

```python
# Set Config
CONFIG = {
    "data_dir": "./gene_vs_sample_MG.csv"
    , "genes": ["HK1", "HK2", "LDHA", "LDHB", "ARHGAP11B"]
    , "color": sns.diverging_palette(255, 10, l=50, s=100, n=40, center="dark")
    , "cbar": {"shrink": .4}
}
```

```python
# Load Data
rawdata = pd.read_csv(CONFIG['data_dir'], index_col=0)
```

```python
# Shape Data
data = np.log2(rawdata+1).T.sort_index().T
data.index.name=None
```

```python
# TPM heatmap
hTPM = data.loc[CONFIG["genes"], :]
plt.figure(figsize=(10, 10))
sns.heatmap(hTPM, cmap=CONFIG["color"], square=True, robust=True, center=hTPM.mean().mean(), cbar=True, cbar_kws=CONFIG["cbar"])
plt.savefig("./HeatMap_TPM_mini.png", facecolor="white", dpi=300, bbox_inches="tight", pad_inches=0.05)
```

```python
# Calculating Z score
Zdata = data.loc[CONFIG["genes"], :].T.apply(zscore).T
```

```python
# Z score heat map
plt.figure(figsize=(10, 10))
sns.heatmap(Zdata, cmap=CONFIG["color"], square=True, robust=True, center=0, cbar=True, cbar_kws=CONFIG["cbar"])
plt.savefig("./HeatMap_Zscore_mini.png", facecolor="white", dpi=300, bbox_inches="tight", pad_inches=0.05)
```
